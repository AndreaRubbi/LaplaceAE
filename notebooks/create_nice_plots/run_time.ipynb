```python
import json
import matplotlib.pyplot as plt
import numpy as np
import os
```

# As function of datasize

### Linear layer network

```python
number_of_layers=5
data_sizes=(10, 50, 100, 500, 1000, 1500, 2000, 2500)
backends=("layer", "row", "la")

performance = {backend : [] for backend in backends}
for data_size in data_sizes:
    for backend in backends:
        path = f"../../figures/run_time/no_conv/{backend}_{data_size}_{number_of_layers}.json"
        if os.path.isfile(path):
            with open(path) as json_file:
                data = json.load(json_file)
                
            performance[backend].append([data_size, data["run_time"], data["memory"]])
            
performance = {backend : np.array(performance[backend]) for backend in backends}
```

```python
for backend in backends:
    data = performance[backend]
    plt.plot(data[:,0], data[:,1], '-o', label=backend)

plt.legend()
plt.xlabel("Input size")
plt.ylabel("Time [sec]")
plt.show()
```

```python
for backend in backends:
    data = performance[backend]
    plt.plot(data[:,0], data[:,2], '-o', label=backend)

plt.legend()
plt.xlabel("Input size")
plt.ylabel("Memory [Mb]")
plt.show()
```
